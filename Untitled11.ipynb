```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Example Confusion Matrix
TP, TN, FP, FN = 3, 4, 0, 1
cm = [[TP, FN],[FP, TN]]
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Correct","Hallucination"],
            yticklabels=["Correct","Hallucination"])
plt.title("Confusion Matrix")
plt.savefig("confusion_matrix.png")
plt.show()

# Performance Metrics Bar Chart
accuracy, precision, recall, f1 = 0.875, 0.9, 0.85, 0.875
labels = ["Accuracy","Precision","Recall","F1"]
values = [accuracy, precision, recall, f1]
plt.figure()
plt.bar(labels, values)
plt.title("Performance Metrics")
plt.ylim(0,1)
plt.savefig("performance_bar_chart.png")
plt.show()

# Live Input Testing Example
text_output = """AI Generated: The sun revolves around the Earth.
Reference: The Earth revolves around the Sun.
Similarity: 0.32
Prediction: Hallucination"""
plt.figure(figsize=(8,2))
plt.text(0.01, 0.5, text_output, fontsize=12)
plt.axis('off')
plt.savefig("live_input_output.png")
plt.show()
```
